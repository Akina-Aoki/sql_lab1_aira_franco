# Restore data in dfs dictionary and Registering 

```python
dfs = {}

with duckdb.connect("data/sakila.duckdb") as conn:
    for name in description['name']:
        if name[0] != '_':
            dfs[name] = conn.sql(f"FROM sakila.{name};").df()

            duckdb.register(name, dfs[name])

dfs['actor'].head()
```
